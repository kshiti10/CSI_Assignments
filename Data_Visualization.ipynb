# Iris Dataset Visualizations
This notebook includes various types of visualizations for the Iris dataset:
- Line Graph
- Histogram
- Pie Chart
- Box & Whiskers Plots

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Sample Data
data = {
    'Year': [2018, 2019, 2020, 2021, 2022],
    'Sales': [250, 270, 300, 280, 320],
    'Region': ['North', 'South', 'East', 'West', 'Central'],
    'Profit': [50, 65, 60, 70, 80]
}
df = pd.DataFrame(data)

# 1. Line Graph
plt.figure(figsize=(8, 4))
plt.plot(df['Year'], df['Sales'], marker='o', color='blue')
plt.title('Sales Over Years')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.grid(True)
plt.tight_layout()
plt.show()

# 2. Histogram
plt.figure(figsize=(8, 4))
plt.hist(df['Profit'], bins=5, color='green', edgecolor='black')
plt.title('Profit Distribution')
plt.xlabel('Profit')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

# 3. Pie Chart
plt.figure(figsize=(6, 6))
plt.pie(df['Sales'], labels=df['Region'], autopct='%1.1f%%', startangle=140)
plt.title('Sales by Region')
plt.tight_layout()
plt.show()

# 4. Box & Whiskers Plot (Correctly using long-form)
df_long = pd.DataFrame({
    'Metric': ['Sales'] * len(df) + ['Profit'] * len(df),
    'Value': df['Sales'].tolist() + df['Profit'].tolist()
})

plt.figure(figsize=(8, 4))
sns.boxplot(x='Metric', y='Value', data=df_long)
plt.title('Boxplot of Sales and Profit')
plt.tight_layout()
plt.show()
```
